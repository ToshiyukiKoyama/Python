```python
#getting apple's historical data from 2010-01-01 to today
df = yf.download("AAPL",start = "2020-01-01", end = date.today())

#create a simple chart
df['Close'].plot(title = "Apple Stock",figsize=(8,5))
plt.show()
```

```python
import pandas_ta as ta
```

**pandas_ta.rsi** refers to the Relative Strength Index (RSI) functionality within the pandas_ta library in Python. The pandas_ta library is an extension of the popular data manipulation library pandas and is designed for technical analysis of financial markets.

**The Relative Strength Index (RSI) is a momentum oscillator that measures the speed and change of price movements.** It is used to identify overbought or oversold conditions in a market, potentially signaling a reversal in price direction. RSI values range from 0 to 100, with readings above 70 typically considered overbought and readings below 30 considered oversold.

**The pandas_ta.cci** function in the pandas_ta library calculates the Commodity Channel Index (CCI), which is a popular momentum indicator used in technical analysis. CCI measures the difference between the current price and its historical average in relation to its normal variability.

```python
df["RSI(2)"] = ta.rsi(df["Close"], length=2)
df["RSI(7)"] = ta.rsi(df["Close"], length=7)
df["RSI(14)"] = ta.rsi(df["Close"], length=14)
df["CCI(10)"] = ta.cci(close = df["Close"], length=10, high = df["High"], low = df["Low"])
df["CCI(100)"] = ta.cci(close = df["Close"], length=100, high = df["High"], low = df["Low"])
```

```python
#dropping na in df
df = df.dropna()
```

```python
#create a plot showing some of our indicators
df.plot(y="RSI(2)")
df.plot(y="CCI(100)")

df.head(5)
```
